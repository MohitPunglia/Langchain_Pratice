```python
!pip install langchain-groq
!pip install python-dotenv
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from google.colab import userdata
from langchain.schema.runnable import RunnableSequence,RunnableParallel,RunnablePassthrough,RunnableLambda
```

```python
model=ChatGroq(model_name='llama3-70b-8192',api_key=userdata.get('GROQ_API_KEY'))
```

```python
template=PromptTemplate(template='Generate a fun fact about {topic}',
               input_variables={'topic'})
```

```python
def word_count(text):
  return len(text.split())
```

```python
parser=StrOutputParser()
```

```python
joke_chain=RunnableSequence(template,model,parser)
```

```python
parallel_chain=RunnableParallel({
    'Joke':RunnablePassthrough(),
    'Word_count':RunnableLambda(word_count)
})
```

```python
chain=RunnableSequence(joke_chain,parallel_chain)
result=chain.invoke({'topic':'IPL'})
```

```python
final_result = """{} \n word count - {}""".format(result['Joke'], result['Word_count'])
print(final_result)
```
